# The Elements of Statistical Learning
## Chapter 2: Overview of Supervised Learning

### Intro

Supervised learning: provided inputs, predict the output

Variable types:
* Quantitative: e.g. glucose level prediction
* Qualitative / categorical: 
     * e.g. species of Iris... Has a finite set, e.g. G = { Virginica, Setosa, Versicolor } 
     * e.g. handwritten digits... e.g. G = { 0, 1, ..., 9 }
     
Prediction tasks:
* Regression > Quantitative
* Classification > Qualitative

 "X" > input variable, or feature
 
 "Y" > output variable
 
 "G" can be used for classification / group

Can loosely state the learning task as follows: 
    
Given the value of an input vector X,
make a good prediction of the output Y,
denoted by $\hat{Y}$. 

If Y takes values in $\mathbb{R}$ (real number range) then so should $\hat{Y}$

### Linear Models and Least Squares

Linear model:
Given a vector of inputs: ${X^T} = ({X_1, X_2, ..., X_p})$

We predict the output Y using the model:

$
{\hat{Y} = \hat{β}_0 + \sum\limits_{j=1}^p X_j  \hat{β}_j}
$

${\hat{β}_0}$ is the intercept (also known as the bias)
