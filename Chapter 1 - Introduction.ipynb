# The Elements of Statistical Learning
## Chapter 1: Introduction

### Intro

**Statistical Learning** examples:
* Predict whether someone will have a heart attack or cancer based on demographics, diet, clinical measurements
* Predict stock price 6 months from now, based on company performance measures and economic data
* Identify numbers in post code, based on training images

Typically, there are 2 types of **learning from data**:
* Outcome: such as stock price
* Categorical: such as spam or ham

We predict these based on a set of **features** (such as diet and clinical measurements).

We use the **training set** of data to create a prediction model.  

**Supervised learning**: observe the outcome of results from training data, and use these to make predictions on new data

**Unsupervised learning**: observe only the features and have no measurements of the outcome
